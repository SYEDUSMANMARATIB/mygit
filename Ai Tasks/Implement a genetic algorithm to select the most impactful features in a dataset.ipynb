```python
import random
import sklearn
from sklearn.model_selection import train_test_split
from sklearn import tree as tree
from sklearn import preprocessing
from sklearn import metrics
import pandas as pd
from pandas import ExcelWriter
from pandas import ExcelFile
import numpy as np
```

```python
data = pd.read_excel('Training_Data.xlsx')
data
```

```python
Columns = []
for i in data:
    Columns.append(i)
length = len(Columns)
```

```python
data[Columns[5]]= data[Columns[5]].fillna('No_DM')
data[Columns[6]]=data[Columns[6]].fillna('No_CKD')
data[Columns[7]]=data[Columns[7]].fillna('No_Anemia')
data[Columns[8]]=data[Columns[8]].fillna('No_depression')
data[Columns[9]]=data[Columns[9]].fillna('No_COPD')
replace_value = data[Columns[4]].mode()
data[Columns[4]]=data[Columns[4]].fillna(replace_value[0])
```

```python
label_encoder = preprocessing.LabelEncoder()
data[Columns[2]]= label_encoder.fit_transform(data[Columns[2]])
data[Columns[3]]= label_encoder.fit_transform(data[Columns[3]])
data[Columns[4]]= label_encoder.fit_transform(data[Columns[4]])
data[Columns[5]]= label_encoder.fit_transform(data[Columns[5]])
data[Columns[6]]= label_encoder.fit_transform(data[Columns[6]])
data[Columns[7]]= label_encoder.fit_transform(data[Columns[7]])
data[Columns[8]]= label_encoder.fit_transform(data[Columns[8]])
data[Columns[9]]= label_encoder.fit_transform(data[Columns[9]])
new_data = data
new_data = new_data.drop(['EncounterId','PatientId'], axis = 1)
```

```python
y = new_data[Columns[length-1]]
X = new_data.drop(Columns[length-1], axis=1)
y = pd.DataFrame(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print("Accuracy:", (metrics.accuracy_score(y_test,y_pred))*100)
```

```python
Columns = []
for i in new_data:
    Columns.append(i)
length = len(Columns)

def fitness_func(arr,new_data,Columns,y):
    fitness = []
    for i in range(0,100):
        df = pd.DataFrame()
        temp = []
        for j in range(0,54):
            if(arr[i][j] == 1):
                extracted_col = new_data[Columns[j]]
                temp.append(extracted_col)
        df = pd.DataFrame(temp)
        df = df.transpose()
        new_Columns = []
        for i in df:
            new_Columns.append(i)
        new_length = len(new_Columns)
        X = df.drop(new_Columns[new_length-1], axis=1)
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
        clf = tree.DecisionTreeClassifier()
        clf = clf.fit(X_train, y_train)
        y_pred = clf.predict(X_test)
        fitness.append(metrics.accuracy_score(y_test,y_pred)*100)
    return fitness
fitness = fitness_func(arr,new_data,Columns,y)
print(fitness)
```
