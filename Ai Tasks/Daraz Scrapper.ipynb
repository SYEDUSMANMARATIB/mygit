```python
n=input("Enter search words: ")

n=n.replace(' ','+')
n="mobile+covers"
link1="https://www.daraz.pk/catalog/?q="
link2="&_keyori=ss&from=input&spm=a2a0e.home.search.go.6a274937IJxjwH"
link=link1+n+link2
print(link)
driver = webdriver.Chrome()
driver.get(link)
#code = requests.get(final_link)
page = soup(driver.page_source, 'html.parser')
# products = page.find('div', {'class': 'c1_t2i'})


Name=[]
Daraz_Mall=[]
Original_Price =[]
Discounted_Price =[]
Rating =[]
Reviews =[]
Seller_Country=[]
Total_Sellers=0
Average_Price=0
Average_Ratings=0
Average_Reviews=0

source = driver.page_source
soup = BeautifulSoup(source)
m=0
for a in soup.findAll(attrs={'class':'c16H9d'}):
    m=m+1


# # Name

# In[158]:


for i in range(1,m+1):
    
    if i==1:
        
        try:
            target_name=driver.find_element_by_xpath('//*[@id="root"]/div/div[2]/div[1]/div/div[1]/div[2]/div[1]/div/div/div[2]/div[2]/a')
            
        except:
            target_name=driver.find_element_by_xpath('//*[@id="root"]/div/div[3]/div[1]/div/div[1]/div[2]/div[1]/div/div/div[2]/div[2]/a')  
            
    else:
        
        try:
            target_name=driver.find_element_by_xpath('//*[@id="root"]/div/div[2]/div[1]/div/div[1]/div[2]/div['+str(i)+']/div/div/div[2]/div[2]/a')
            
        except:
            target_name=driver.find_element_by_xpath('//*[@id="root"]/div/div[3]/div[1]/div/div[1]/div[2]/div['+str(i)+']/div/div/div[2]/div[2]/a')          
            
    Name.append(target_name.text)


# # Daraz Mall

# In[159]:


d=soup.find_all('div', class_='c3vCyH')
k="""<div class="c3vCyH"><i class="ic-dynamic-badge ic-dynamic-badge-lazMall ic-dynamic-group-1" style='background-image: url("//laz-img-cdn.alicdn.com/tfs/TB1rYuurhTpK1RjSZFMXXbG_VXa-205-48.png"); width: 68.3333px; height: 16px;'></i></div>"""

for i in d:
    if(str(i) == k):
            Daraz_Mall.append("Yes")
    else:
        Daraz_Mall.append("No") 


# # Original Price And Discounted Price

# In[160]:


for i in range(1,m+1):
    try:
        target_dprice=driver.find_element_by_xpath('//*[@id="root"]/div/div[2]/div[1]/div/div[1]/div[2]/div['+str(i)+']/div/div/div[2]/div[4]/span[1]/del')
    except:
        target_dprice=driver.find_element_by_xpath('//*[@id="root"]/div/div[2]/div[1]/div/div[1]/div[2]/div['+str(i)+']/div/div/div[2]/div[3]/span')
    
#     target_dprice=driver.find_element_by_xpath('//*[@id="root"]/div/div[2]/div[1]/div/div[1]/div[2]/div['+str(i)+']/div/div/div[2]/div[3]/span')
    
#     Original_Price.append(target_oprice.text)
    Discounted_Price.append(target_dprice.text)


# # Average Price

# In[161]:


sum=0
a=0
for i in Original_Price:
    x = i.replace("Rs.", "")
    x = x.replace(",", "")
    x = x.replace(" ", "")
    sum=sum+int(x)
    a=a+1
if a > 0:
    Average_Price=sum/a
else:
    Average_Price=0


# # Reviews

# In[162]:


for i in range(1,m+1):
    try:
        target_reviews=driver.find_element_by_xpath('//*[@id="root"]/div/div[2]/div[1]/div/div[1]/div[2]/div['+str(i)+']/div/div/div[2]/div[5]/div/span')
        Reviews.append(target_reviews.text)
    except:
        target_reviews="0"
        Reviews.append(target_reviews)


# # Average Reviews

# In[163]:


sum=0
a=0
for i in Reviews:
    x = i.replace("(", "")
    x = x.replace(")", "")
    x = x.replace(" ", "")
    x = x.replace("NILL", "0")
    sum=sum+int(x)
    a=a+1
if a > 0:
    Average_Reviews=sum/a
else:
    Average_Reviews=0


# # Location

# In[164]:


for i in range(1,m+1):
    target_loc=driver.find_element_by_xpath('//*[@id="root"]/div/div[2]/div[1]/div/div[1]/div[2]/div['+str(i)+']/div/div/div[2]/div[5]/span')
    Seller_Country.append(target_loc.text)


# # Rating

# In[165]:


a=0
sum=0
target_rating = soup.find_all("div", {"class" : "c2prKC"})
for i in range(0,(len(target_rating))):
    value = target_rating[i]
    rating = value.find('div', class_ = 'c15YQ9')
    strng =str(rating)
    two =strng.count("c3dn4k cbDGcO")
    three =strng.count("c3dn4k c3fsPU")
    four = strng.count("c3dn4k c1e2gb")
    five =strng.count("c3dn4k c3An30")
    six = strng.count("c3dn4k c3DcGB")
    seven =strng.count("c3dn4k c1wCjy")
    eight =strng.count("c3dn4k c17YMy")
    nine =strng.count("c3dn4k cF1vkb")
    full=strng.count("c3dn4k c3EEAg")
    two*=0.2
    three*=0.3
    four*=0.4
    five*=0.5
    six*=0.6
    seven*=0.7
    eight*=0.8
    nine*=0.9
    ratings=two+three+four+five+six+seven+eight+nine+full
    sum=ratings+sum
    a=a+1
    Rating.append(ratings)
Average_Rating=sum/a

df = pd.DataFrame(Name,columns=[' Name']) 
df['Price'] = Discounted_Price
df[' Rating'] = Rating
df[' Reviews'] = Reviews
print(df)
driver.quit()
```
