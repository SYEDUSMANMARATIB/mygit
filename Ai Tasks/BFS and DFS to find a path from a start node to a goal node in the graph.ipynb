```python
class Graph:
    def __init__(self,max_v):
        self.max_v= max_v
        self.G={i:[] for i in range (self.max_v)}
        
    def init_graph(self,nodes):
        self.G={i: [] for i in nodes}
        
    def add_Edge(self,src,dest):
        if dest not in self.G[src]:
            self.G[src].append(dest)
            self.G[src].append(src)
    def print_Graph(self):
        for i in self.G:
            print(i,self.G[i])
    def BFS (self,start,goal):
        Q=[start]
        
        visited=[start]
        while Q:
            
            vertex = Q.pop(0)
            
            if vertex == goal:
                print ('reached')
                
                return goal
            
            for i in self.G[vertex]:
                if i  not in visited:
                    visited.append(i)
                    
                    Q.append(i)
                    
    def DFS (self,start,goal):
        A=[start]
        
        visited=[start]
        while A:
            
            vertex = A.pop()
            
            if vertex == goal:
                print ('reached')
                
                return goal
            
            for i in self.G[vertex]:
                if i  not in visited:
                    visited.append(i)
                    
                    A.append(i)                    
        
    
```

```python
def gen(state, m, b):
    temp = state[:]
    if m == 'l': temp[b], temp[b - 1] = temp[b - 1], temp[b]
    if m == 'r': temp[b], temp[b + 1] = temp[b + 1], temp[b]
    if m == 'u': temp[b], temp[b - 3] = temp[b - 3], temp[b]
    if m == 'd': temp[b], temp[b + 3] = temp[b + 3], temp[b]
    return temp

def possible_moves(state, visited_states):
    b = state.index(-1)
    d = []
    pos_moves = []
    if b <= 5: d.append('d')
    if b >= 3: d.append('u')
    if b % 3 > 0: d.append('l')
    if b % 3 < 2: d.append('r')
    for i in d:
        temp = gen(state, i, b)
        if not temp in visited_states: pos_moves.append(temp)

    return pos_moves
```
