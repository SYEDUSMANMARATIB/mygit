```python
##col = [ 'price','area','length','width','bedrooms','bathrooms','stories','mainroad','guestroom','basement','hotwaterheating','airconditioning','parking','prefarea','furnishingstatus']
df = pd.read_csv('Housing.csv',sep=',')
##df.head()
df.mainroad.map(dict(yes=1, no=0))
df = df.replace( to_replace = {"no" : 0, "yes" : 1})
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
X = df.drop('furnishingstatus',axis=1)

import pandas as pd
from sklearn import preprocessing

x = X.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
X = pd.DataFrame(x_scaled)

y = df[['furnishingstatus']]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_clf = RandomForestClassifier(criterion='gini')   
rf_clf.fit(X_train,y_train)
```

```python
RandomForestClassifier(criterion='gini')
```

```python
y_predict = rf_clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
accuracy_score(y_test,y_predict)
```

```python
cities=['A','B','C','D','E']
samp=[]

for i in range(len('cities')-1):
    for j in range(len('cities')-1):
        for k in range(len('cities')-1):
            samp.append(cities[i]+cities[j]+cities[k])
```

```python
zoo=0
for i in range(0, len(samp)):
    if samp[i][0] == samp[i][1] or samp[i][0] == samp[i][2] or samp[i][1] == samp[i][2]:
        zoo+=1
        
```

```python
print(zoo)
```

```python
print("The probability of 2 people be in a same city are  :",zoo/125)
```

```python
print("The other probability is   :",(125-zoo)/125)
```
