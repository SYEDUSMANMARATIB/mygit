## Task : 3

```python
picture2 = cv2.imread('picture2.jpg', 0)
picture2 = cv2.resize(picture2, dim)
cv2_imshow(picture2)
```

```python
img = picture2[100:160, 150:200]
cv2_imshow(img)
```

```python
picture2 = deepcopy(picture2)
picture2[100:160, 100:150] = img
picture2[170:230, 100:150] = img
picture2[40:100, 80:130] = img
cv2_imshow(picture2)
```

```python
def put_image(org, small, x, y, w, h):
    org2 = deepcopy(org)
    org2[x:x+w, y:y+h] = small
    return org2

cv2_imshow(put_image(picture2, img, 220, 20, 60, 50))
cv2_imshow(put_image(picture2, img, 180, 30, 60, 50))
cv2_imshow(put_image(picture2, img, 230, 40, 60, 50))
```

```python
picture2 = deepcopy(picture2)
picture2[40:100, 80:130] = (picture2[40:100, 80:130] + img)/2
cv2_imshow(picture2)
```
